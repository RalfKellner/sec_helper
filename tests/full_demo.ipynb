```python
apple_eightks = apple_filing_history[apple_filing_history.form == "8-K"]
apple_eightks = apple_eightks[apple_eightks.filingDate > "2020-01-01"]
apple_eightks.reset_index(drop = True, inplace = True)

apple_eightks.loc[:, "8K_keys"] = ""
apple_eightks.loc[:, "8K_text"] = ""

for idx, row in apple_eightks.iterrows():
    apple_eightk_keys, apple_eightk_filing = get_8K_filing_and_exhibits(
        apple_cik,
        "user@uni-passau.de",
        accession_number = row["accessionNumber"],
        min_length_extract_from_body=100,
        raw_html=False,
        separator_character="-;-",
        return_keys=True   
    )
    print(idx)

    apple_eightks.loc[idx, "8K_keys"] = apple_eightk_keys
    apple_eightks.loc[idx, "8K_text"] = apple_eightk_filing
```

```python
import pandas as pd
for idx, row in apple_eightks.iterrows():
    try:
        keys = row["8K_keys"].split("-;-")[:-1]
        filing_extended_tmp = pd.DataFrame([row.drop(["8K_keys", "8K_text"])] * len(keys))
        filing_extended_tmp.loc[:, "8K_key"] = keys
        filing_extended_tmp.loc[:, "text"] = row["8K_text"].split("-;-")[:-1]
        filings_extended = pd.concat((filings_extended, filing_extended_tmp), ignore_index=True)
    except:
        keys = row["8K_keys"].split("-;-")[:-1]
        filing_extended_tmp = pd.DataFrame([row.drop(["8K_keys", "8K_text"])] * len(keys))
        filing_extended_tmp.loc[:, "8K_key"] = keys
        filing_extended_tmp.loc[:, "text"] = row["8K_text"].split("-;-")[:-1]
        filings_extended = filing_extended_tmp.copy()
```
