```python
from sec_helper.sec_helper_functions import cik_company_tickers

ciks_and_companies = cik_company_tickers("user@uni-passau.de", deduplicate=True)
ciks_and_companies.head()
```

```python
from sec_helper.sec_helper_functions import get_filing_history

apple_cik = ciks_and_companies[ciks_and_companies.ticker == "AAPL"].cik.values[0]
apple_filing_history = get_filing_history(apple_cik, "user@uni-passau.de")
apple_filing_history.head()
```

```python
from sec_helper.sec_helper_functions import get_10_form_filing

apple_tenks = apple_filing_history[apple_filing_history.form == "10-K"]

apple_tenk_filing = get_10_form_filing(
    apple_cik,
    "user@uni-passau.de",
    accession_number = apple_tenks["accessionNumber"].iloc[0],
    primary_doc = apple_tenks["primaryDocument"].iloc[0],
    min_length_extract_from_body = 2000,
    raw_html = False
)

apple_tenk_filing.split(".")[100:200]
```

```python
from sec_helper.sec_helper_functions import get_10_form_filing

apple_tenqs = apple_filing_history[apple_filing_history.form == "10-Q"]

apple_tenq_filing = get_10_form_filing(
    apple_cik,
    "user@uni-passau.de",
    accession_number = apple_tenqs["accessionNumber"].iloc[0],
    primary_doc = apple_tenqs["primaryDocument"].iloc[0],
    min_length_extract_from_body = 1000,
    raw_html = False
)

apple_tenq_filing.split(".")[100:200]
```

```python
from sec_helper.sec_helper_functions import get_8K_filing_and_exhibits

apple_eightks = apple_filing_history[apple_filing_history.form == "8-K"]

apple_eightk_filing = get_8K_filing_and_exhibits(
    apple_cik,
    "user@uni-passau.de",
    accession_number = apple_eightks["accessionNumber"].iloc[4],
    min_length_extract_from_body=100,
    raw_html=False,
    separator_character="-;-"   
)

apple_eightk_filing.split("-;-")
```
